```python
# import required library 
import camelot as cl
import pandas as pd
import numpy as np
```

```python
# Read PDF file and extract table form it and convert it to pandas data frame
data = cl.read_pdf("assest/test_input.pdf", flavor='stream')
df = data[0].df
```

```python
df.columns = df.iloc[1]
df = df.drop(1)
```

```python
# Creat Currancy column
temp = df['Value date'].str.extract('Deposits (.*)')
df['Currency'] = np.where(temp.isna(), temp.ffill(), '')
```

```python
def debit(val):
    if val.startswith('('):
        return val
    else:
        return ''

def credit(val):
    if val.startswith('('):
        return ''
    else:
        return val
```

```python
# Creat Debit and Credit Column with the help of debit and credit functions
df['Debit'] = df['Amount'].apply(debit).replace({',': '', '\(':'', '\)':''}, regex=True)
df['Credit'] = df['Amount'].apply(credit).replace({',': ''}, regex=True)
```

```python
# Drop Ammount column
df.drop('Amount', inplace=True, axis=1)
```

```python
df.head()
```

```python
df['Description'][7] = df['Description'][7] + df['Description'][8]
df['Description'][9] = df['Description'][9] + df['Description'][10]
df['Description'][20] = df['Description'][20] + df['Description'][21]
df['Description'][22] = df['Description'][22] + df['Description'][23]
```

```python
nan_value = float("NaN")
df['Order date'].replace("", nan_value, inplace = True)
df = df.dropna(subset = ['Order date'])
```

```python
df.head()
```

```python
# Converting Format of date time
df['Value date'] = pd.to_datetime(df['Value date'], format='%Y%m%d')
df['Order date'] = pd.to_datetime(df['Order date'], format='%Y%m%d')
```

```python
# Saving data frame to excel format 
df.to_excel("file.xlsx", index = False)
```
